```python
premium_users_count = (data['is_premium'] == 1).sum()
premium_users_count
```

```python
non_premium_users_count = (data['is_premium'] == 0).sum()
non_premium_users_count
```

```python
labels = ['Premium', 'Non-Premium']
counts = [premium_users_count, non_premium_users_count]
plt.pie(counts, labels=labels, autopct='%0.2f%%')
plt.title('Premium vs Non-Premium Users')
plt.show()
```

```python
ques_asked_by_faang = (data['asked_by_faang'] == 1).sum()
ques_asked_by_faang
```

```python
ques_notasked_by_faang = (data['asked_by_faang'] == 0).sum()
ques_notasked_by_faang
```

```python
labels = ['ques_asked_by_faang', 'ques_not_asked_by_faang']
counts = [ques_asked_by_faang, ques_notasked_by_faang]
plt.pie(counts, labels=labels, autopct='%0.2f%%')
plt.title('ques_asked_by_faang vs ques_not_asked_by_faang')
plt.show()
```

```python
total_question_titles = len(data['title'])
print("Total questions:", total_question_titles)
```

```python
easy_questions_count = (data['difficulty'] == 'Easy').sum()
medium_questions_count = (data['difficulty'] == 'Medium').sum()
hard_questions_count = (data['difficulty'] == 'Hard').sum()
print("Number of easy questions:", easy_questions_count)
print("Number of medium questions:", medium_questions_count)
print("Number of hard questions:", hard_questions_count)
```

```python
labels = ['Easy', 'Medium','Hard']
counts = [easy_questions_count, medium_questions_count, hard_questions_count]
plt.pie(counts, labels=labels, autopct='%0.2f%%')
plt.title('Easy vs Medium vs Hard')
plt.show()
```

```python
plt.bar(labels, counts, color=['blue', 'orange', 'green'])
plt.xlabel('Difficulty')
plt.ylabel('Count')
plt.title('Count of Easy vs Medium vs Hard Questions')
plt.show()
```

```python
acceptance_rate_list = data['acceptance_rate'].tolist()
bins = [0, 40, 80, 100]

plt.hist(acceptance_rate_list, bins=bins,edgecolor='black')
plt.xlabel('Acceptance Rate Range')
plt.ylabel('Number of Questions')
plt.title('Number of Questions vs Acceptance Rate Ranges')
plt.show()
```

```python
avg_acceptance_count = (data['acceptance_rate']).mean()
avg_acceptance_count
```

```python

empty_solution_links_count = data['solution_link'].isnull().sum()
solution_links_count = data['solution_link'].notnull().sum()
print("Questions with empty solution links:", empty_solution_links_count)
print("Questions with solution links:", solution_links_count)
```

```python
max_freq_index = data['frequency'].idxmax()
question_with_max_freq = data.loc[max_freq_index, 'description']
print("Question with max frequency:",question_with_max_freq)
```

```python
min_disc_index = data['discuss_count'].idxmin()
question_with_min_disc = data.loc[min_disc_index, 'description']
print("Question with minimum discuss count:",question_with_min_disc)
```

```python
data['accepted'] = [str(x).strip('MK') for x in data['accepted']]
data['accepted'] = data['accepted'].replace({'M': 'e6', 'K': 'e3'}, regex=True).map(lambda x: float(x) if isinstance(x, str) else x)
max_accepted_index = data['accepted'].idxmax()
question_with_max_accepted_rate = data.loc[max_accepted_index, 'description']
accepted_rate_of_max_question = data.loc[max_accepted_index, 'accepted']
print("Question with the highest accepted rate:", question_with_max_accepted_rate)
print("Accepted rate of that question:", accepted_rate_of_max_question)
```

```python
data['submissions'] = [str(x).strip('MK') for x in data['submissions']]
data['submissions'] = data['submissions'].replace({'M': 'e6', 'K': 'e3'}, regex=True).map(lambda x: float(x) if isinstance(x, str) else x)
max_submission_index = data['submissions'].idxmax()
question_with_max_submission_rate = data.loc[max_submission_index, 'description']
submission_rate_of_max_question = data.loc[max_submission_index, 'submissions']
print("Question with the highest submission rate:", question_with_max_submission_rate)
print("Submission rate of that question:", submission_rate_of_max_question)
```

```python
missing_values = data['companies'].isna().any()
if missing_values:
    data['companies'].fillna('', inplace=True)
company_lists = data['companies'].str.split(',')
all_companies = [company for sublist in company_lists for company in sublist]
unique_companies = sorted(set(all_companies))

print("Unique companies (alphabetical order):")
for company in unique_companies:
    print(company)

num_unique_companies = len(unique_companies)
print("Number of unique companies:", num_unique_companies)
```

```python
import pandas as pd
missing_values = data['companies'].isna().any()
if missing_values:
    data['companies'].fillna('', inplace=True)

all_companies = [company for sublist in data['companies'].str.split(',') for company in sublist]
company_counts = pd.Series(all_companies).value_counts()
print("Company occurrences:")
with pd.option_context('display.max_rows', None):
    print(company_counts)
```

```python
import pandas as pd
data.dropna(subset=['related_topics'], inplace=True)
data = data[data['related_topics'] != '']
all_topics = [topic for sublist in data['related_topics'].str.split(',') for topic in sublist]
topics_counts = pd.Series(all_topics).value_counts()
topics_counts = topics_counts[topics_counts.index != '']
print("Related_topics occurrences:")
with pd.option_context('display.max_rows', None):
    print(topics_counts)
```

```python

plt.hist(data['acceptance_rate'], bins=13, color='skyblue', edgecolor='blue')
plt.xlabel('Acceptance Rate')
plt.ylabel('Frequency')
plt.title('Frequency Distribution of Acceptance Rate')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
top_10_companies = company_counts.head(10)
plt.figure(figsize=(9, 5))
plt.bar(top_10_companies.index, top_10_companies.values)
plt.xlabel('Company')
plt.ylabel('Occurrence')
plt.title('Top 10 Company Occurrences')
plt.xticks(rotation=45)  
plt.show()
```
